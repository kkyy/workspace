```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.offline import plot, iplot
```

```python
spx = pd.read_csv('CIC_0503/SPX.csv',index_col=0)
trades = pd.read_csv('CIC_0503/trades.csv',index_col=0)
trades_ES = pd.read_csv('CIC_0503/trades_ES.csv',index_col=0)
trades_Cash = pd.read_csv('CIC_0503/trades_Cash.csv',index_col=0)
```

```python
spx.head()
```

```python
trades.head()
```

```python
a = trades['Notional'].to_list()
b = [float(i.replace(",", "")) for i in a]
c1 = [i*2 / 100000000 for i in b]
trades['Notional_e'] = trades['Notional'].str.replace(',', '').astype(float)/100000000
trades['Notional_e'] = trades['Notional_e'].round(2).astype(str)

a = trades_ES['Notional'].to_list()
b = [float(i.replace(",", "")) for i in a]
c2 = [i*6 / 100000000 for i in b]
trades_ES['Notional_e'] = trades_ES['Notional'].str.replace(',', '').astype(float)/100000000
trades_ES['Notional_e'] = trades_ES['Notional_e'].round(2).astype(str)

a = trades_Cash['Notional'].to_list()
b = [float(i.replace(",", "")) for i in a]
c3 = [i*6 / 100000000 for i in b]
trades_Cash['Notional_e'] = trades_Cash['Notional'].str.replace(',', '').astype(float)/100000000
trades_Cash['Notional_e'] = trades_Cash['Notional_e'].round(2).astype(str)
```

```python
# 
fig = go.Figure()

textpositions = ['top center', 'bottom left', 'top center', 'top left', 'top center','top center','top center','top center']

trace_spx = go.Scatter(x=spx.index, y=spx['SPX'], mode='lines', marker=dict(color='blue', size=1),name='SPX')
trace_trades_ES = go.Scatter(x=trades_ES['Time'], y=trades_ES['SPX'], 
                            mode='markers+text',
                            #text=trades['Notional_e']+'/ SPX:'+trades['SPX'],
                            text=trades_ES['Notional_e'],
                            textposition=textpositions*3,
                            marker_symbol='circle-dot',
                            marker_line_width=2,
                            marker_line_color='rgba(0, 10, 0, 0.7)',
                            marker_color='rgba(0, 255, 0, 20)',
                            marker_size=c2, 
                            name='trades_ES')

trace_trades_Cash = go.Scatter(x=trades_Cash['Time'], y=trades_Cash['SPX'], 
                            mode='markers+text',
                            #text=trades['Notional_e']+'/ SPX:'+trades['SPX'],
                            text=trades_Cash['Notional_e'],
                            textposition=textpositions*3,
                            marker_symbol='circle-dot',
                            marker_line_width=2,
                            marker_line_color='rgba(0, 10, 0, 0.7)',
                            marker_color='rgba(255, 0, 0, 20)',
                            marker_size=c3, 
                            name='trades_Cash')
#

# 
fig.add_traces(trace_spx)
fig.add_traces(trace_trades_Cash)
fig.add_traces(trace_trades_ES)

fig.update_layout(
    #title='SPX',
    xaxis=dict(title="Time",
               #tickmode='linear',
               dtick=180,),
    yaxis=dict(title="SPX"),
    xaxis_tickangle=-45,
    xaxis_tickfont_size=11,
    width=1000,
    height=500,
    showlegend=False,
    template="plotly_white")#ggplot2seaborn
    

# 
iplot(fig)
```
