```python
def get_date_list(begin_date, end_date):
    dates = []
    # Get the time tuple : dt
    dt = datetime.strptime(begin_date, "%Y-%m-%d")
    date = begin_date[:]
    while date <= end_date:
        dates.append(date)
        dt += timedelta(days=1)
        date = dt.strftime("%Y-%m-%d")
    return dates

def get_workdate_list(begin_date, end_date):
    dates = []
    # Get the time tuple : dt
    dt = datetime.strptime(begin_date, "%Y-%m-%d")
    date = begin_date[:]
    while date <= end_date:
        if dt.strftime("%w") in ["1", "2", "3", "4", "5"]:
            dates.append(date)
        dt += timedelta(days=1)
        date = dt.strftime("%Y-%m-%d")
    return dates
```
